```python
import numpy as np
import pandas as pd
import os
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
```

```python
df=pd.read_csv('Downloads\Automobile_data.csv')
df
```

```python
df=df.replace('?',0)
```

```python
df['normalized-losses']=pd.to_numeric(df['normalized-losses'])
```

```python
sns.distplot(df['engine-size'])
plt.show()
```

```python
sns.distplot(df['city-mpg'])
plt.show()
```

```python
sns.distplot(df['normalized-losses'],kde=False,rug=True)
plt.show()
```

```python
df['horsepower']= pd.to_numeric(df['horsepower'])
```

```python
sns.jointplot( df['engine-size'],df['horsepower'])
plt.show()
```

```python
sns.jointplot(df['engine-size'],df['horsepower'],kind="hex")
plt.show()
```

```python
sns.jointplot(df['engine-size'],df['horsepower'],kind="resid")
```

```python
sns.jointplot(df['engine-size'],df['horsepower'],kind="reg")
```

```python
sns.jointplot(df['engine-size'],df['horsepower'],kind="kde")
```

```python
sns.pairplot(df[['normalized-losses','engine-size','horsepower']])
```

```python
sns.stripplot(df['fuel-type'],df['horsepower'])
```

```python
sns.swarmplot(df['fuel-type'],df['horsepower'])
```

```python
sns.boxplot(df['num-of-doors'],df['horsepower'])
```

```python
sns.stripplot(df['fuel-type'],df['horsepower'],jitter=True)
```

```python
sns.boxplot(df['num-of-doors'],df['horsepower'],hue=df['fuel-type'])
```

```python
sns.barplot(df['body-style'],df['horsepower'])
```

```python
sns.barplot(df['body-style'],df['horsepower'],hue=df['engine-location'])
```

```python
sns.countplot(df['body-style'])
```

```python
sns.countplot(df['body-style'],hue=df['engine-location'])
```

```python
sns.pointplot(df['fuel-system'],df['horsepower'])
```

```python
sns.pointplot(df['fuel-system'],df['horsepower'],hue=df['num-of-doors'])
```
